```python
# Bibliotecas
import random
import time
import matplotlib.pyplot as plt
```

```python
n = 20000
lista_ids = random.sample(range(1, 100000), n)
ids_busca = random.sample(lista_ids, 100)
```

```python
# Busca Sequencial
def busca_sequencial(lista, alvo):
    for item in lista:
        if item == alvo:
            return True
    return False
```

```python
def insertion_sort(lista):
    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1

        while j >= 0 and lista[j] > chave:
            lista[j + 1] = lista[j]
            j -= 1

        lista[j + 1] = chave
```

```python
# Busca Binária
def busca_binaria(lista, alvo):
    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == alvo:
            return True
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1

    return False
```

```python
# Estratégia A
# 100 buscas sequenciais
inicio_a = time.time()

for alvo in ids_busca:
    busca_sequencial(lista_ids, alvo)

tempo_a = time.time() - inicio_a
```

```python
# Estratégia B
# Ordenar + 100 buscas binárias
lista_copia = lista_ids.copy()

inicio_b = time.time()

insertion_sort(lista_copia)

for alvo in ids_busca:
    busca_binaria(lista_copia, alvo)

tempo_b = time.time() - inicio_b
```

```python
# Tempos
print("Tempo Estratégia A:", tempo_a)
print("Tempo Estratégia B:", tempo_b)
```

```python
# Gráfico
estrategias = ["Sequencial", "Ordenar + Binária"]
tempos = [tempo_a, tempo_b]

plt.bar(estrategias, tempos)
plt.ylabel("Tempo (segundos)")
plt.title("Comparação de Estratégias de Busca")
plt.show()
```
