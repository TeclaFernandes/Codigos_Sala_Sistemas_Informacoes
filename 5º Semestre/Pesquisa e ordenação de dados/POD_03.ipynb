```python
import random
import time
import matplotlib.pyplot as plt
```

```python
def Busca_linear(lista, alvo):
  for i in range(len(lista)):
    if lista[i] == alvo:
      return i
  return -1
```

```python
tamanhos = [100000, 1000000, 5000000, 10000000]
alvo = -1
resultados = []
```

```python
for n in tamanhos:

  print(f"Testando lista de tamanho {n}")

  lista = random.sample(range(n * 2), n)

  tempoInicio = time.time()
  Busca_linear(lista, alvo)
  tempoFim = time.time()

  tempo = tempoFim - tempoInicio
  resultados.append(tempo)

  print(f"Tempo: {tempo:.6f} segundos")
```

```python
plt.plot(tamanhos, resultados, marker='o')
plt.title("Tempo de busca linear vs Tamanho da lista")
plt.xlabel("Tamanho da lista")
plt.ylabel("Tempo (Segundos)")
plt.grid(True)

plt.show()

print("Resultados: ", resultados)
```
