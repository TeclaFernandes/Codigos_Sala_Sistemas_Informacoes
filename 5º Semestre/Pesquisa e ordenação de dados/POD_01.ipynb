Importando as dependências

```python
import random
import time
import matplotlib.pyplot as plt
```

Função para a busca linear

```python
def busca_linear(lista, alvo):
    for i in range(len(lista)):
        if lista[i] == alvo:
            return i
    return -1
```

Parâmetros do experimento

```python
tamanhos = [1000, 5000, 10000, 20000, 40000, 80000]
repeticoes = 10  # número de vezes que repe cada busca
tempos_medios = []
```

```python
for n in tamanhos:
    lista = random.sample(range(n * 2), n)
    alvo = random.choice(lista)

    tempos = []
    for _ in range(repeticoes):
        inicio = time.time()
        busca_linear(lista, alvo)
        fim = time.time()
        tempos.append(fim - inicio)

    tempo_medio = sum(tempos) / repeticoes
    tempos_medios.append(tempo_medio)
    print(f"Tamanho: {n} | Tempo médio: {tempo_medio:.8f} segundos")
```

Gerando gráficos

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.barh([str(n) for n in tamanhos], tempos_medios, color='skyblue')
ax.set_xlabel("Tempo Médio de Execução (segundos)", fontsize=16)
ax.set_ylabel("Tamanho do Array", fontsize=16)
ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.subplots_adjust(top=0.85, bottom=0.15, left=0.15, right=0.95)
fig.text(0.5, 0.90, "Pesquisa e Ordenação de Dados", ha='center', fontsize=16, fontweight='bold')
fig.text(0.5, 0.87, "Tempo Médio da Busca Linear em Arrays Não Ordenados, 13 de Fevereiro de 2026",
         ha='center', fontsize=12, fontweight='normal')

plt.show()
```
